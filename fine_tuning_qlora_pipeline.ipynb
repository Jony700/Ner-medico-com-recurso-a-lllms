```python
import pandas as pd




def load_and_rename_excel_sheets(file_path, sheet_name_mapping):
  """
  Loads data from specific sheets of an Excel file into pandas DataFrames
  and renames the DataFrames in the output dictionary.

  Args:
    file_path (str): The path to the Excel file.
    sheet_name_mapping (dict): A dictionary where keys are the sheet names
                                in the Excel file and values are the desired
                                names for the DataFrames.

  Returns:
    dict: A dictionary where keys are the desired DataFrame names and values
          are the corresponding pandas DataFrames. Returns an empty dictionary
          if the file is not found or none of the specified sheets exist.
  """
  loaded_dataframes = {}
  try:
    # Use pandas.ExcelFile to open the Excel file
    xlsx = pd.ExcelFile(file_path)
    # Iterate through the sheet name mapping
    for excel_sheet_name, desired_df_name in sheet_name_mapping.items():
      print(excel_sheet_name)
      if excel_sheet_name in xlsx.sheet_names:
        # Read the data from the current sheet into a DataFrame
        df = xlsx.parse(excel_sheet_name)
        # Store the DataFrame in the dictionary with the desired name as the key
        loaded_dataframes[desired_df_name] = df
      else:
        print(f"Warning: Sheet '{excel_sheet_name}' not found in the Excel file.")

    # Close the Excel file
    xlsx.close()

    return loaded_dataframes

  except FileNotFoundError:
    print(f"Error: File not found at '{file_path}'")
    return {}
  except Exception as e:
    print(f"An error occurred: {e}")
    return {}

# Example usage:
excel_file = 'TodasVersoesICD10CMPCS_2017_2025_v1_caracteres_25_02_2025.xlsx'  # Replace with the actual path to your Excel file
sheet_name_mapping = {
    'ICD10CM': 'condições medicas',
    'ICD10PCS': 'procedimentos medicos'
}

renamed_dataframes = load_and_rename_excel_sheets(excel_file, sheet_name_mapping)

# Now you can access the DataFrames using the desired names
if renamed_dataframes:
  if 'condições medicas' in renamed_dataframes:
    df_condicoes_medicas = renamed_dataframes['condições medicas']
    print("DataFrame for 'condições medicas':")
    print(df_condicoes_medicas.head())
    print("-" * 30)
  else:
    print("DataFrame 'condições medicas' not loaded.")

  if 'procedimentos medicos' in renamed_dataframes:
    df_procedimentos_medicos = renamed_dataframes['procedimentos medicos']
    print("DataFrame for 'procedimentos medicos':")
    print(df_procedimentos_medicos.head())
    print("-" * 30)
  else:
    print("DataFrame 'procedimentos medicos' not loaded.")
else:
  print("No dataframes were loaded.")
df_condicoes_medicas.head()
```

```python


# --- Helper Functions ---

def load_dataframe_parquet(file_path):
    """Loads a pandas DataFrame from a Parquet file."""
    try:
        loaded_df = pd.read_parquet(file_path)
        print(f"DataFrame loaded from: {file_path}")
        return loaded_df
    except FileNotFoundError:
        print(f"Error: File not found at '{file_path}'")
        return None
    except Exception as e:
        print(f"Error loading DataFrame from Parquet: {e}")
        return None

def normalize_text(text):
    """Normalizes text: lowercase, strip whitespace, handle multiple spaces."""
    if isinstance(text, str):
        # Lowercase, strip leading/trailing whitespace, replace multiple spaces with single
        return ' '.join(text.lower().strip().split())
    return "" # Return empty string for non-string input
```

```python

import pandas as pd
import random
from collections import Counter
import math

# Placeholder definitions if not provided by the user's environment
if 'normalize_text' not in globals():
    def normalize_text(text):
        """Simple text normalization."""
        if isinstance(text, str):
            return text.strip().lower()
        return ""

if 'loaded_dfs' not in globals():
    # Create dummy dataframes for demonstration
    print("Creating dummy dataframes for demonstration.")
    loaded_dfs = {
        'condicoes': pd.DataFrame({
            'Descrição PT_(Longa)': ['Dor de cabeça', 'Febre', 'Fratura do fêmur', 'Queda de escada', 'Exposição a fumo'],
            'Secção ICD-10-CM_Código': ['R51', 'R50', 'S72', 'W00', 'Z57'],
            'Capitulo ICD-10-CM_ Código': ['R', 'R', 'S', 'W', 'Z']
        }),
        'medicamentos': pd.DataFrame({
            'Nome do medicamento': ['Paracetamol 500mg', 'Ibuprofeno', 'Amoxicilina 250mg/5ml', 'Paracetamol'],
            'Substância Ativa/DCI': ['Paracetamol', 'Ibuprofeno', 'Amoxicilina', 'Paracetamol'],
            'Dosagem': ['500mg', '400mg', '250mg/5ml', '750mg']
        }),
        'procedimentos': pd.DataFrame({
            'Descrição Linguagem Corrente': ['Cirurgia de apendicite', 'Radiografia de tórax', 'Consulta médica']
        }),
        'allergens': pd.DataFrame({
            'Preferred Term (PT) PT | Termo de Preferência PT': ['Pólen', 'Ácaros', 'Penicilina']
        })
    }

if 'VALID_CLASSES' not in globals():
    VALID_CLASSES = ['M', 'DS', 'D', 'S', 'C', 'PM', 'AG']
    print(f"Using default VALID_CLASSES: {VALID_CLASSES}")

# Define the total number of samples for the small evaluation set
TOTAL_EVALUATION_SAMPLES = 100

# --- Prepare Evaluation Data (no IDs, no linking) ---
print("Preparing data for evaluation and training...")

# Ensure VALID_CLASSES includes all expected labels
if not all(cls in VALID_CLASSES for cls in ['M','DS','D','S','C','PM','AG']):
     print("Warning: VALID_CLASSES might not include all expected labels.")
     # Attempt to add missing ones if possible, or raise error
     # For robustness, let's ensure required classes are in VALID_CLASSES
     required_classes = ['M','DS','D','S','C','PM','AG']
     for cls in required_classes:
         if cls not in VALID_CLASSES:
             print(f"Adding missing class '{cls}' to VALID_CLASSES.")
             VALID_CLASSES.append(cls)


# Initialize term containers for each class exactly once
all_terms_by_class = {cls: set() for cls in VALID_CLASSES}

# --- Load and process conditions for D, S, C ---
loaded_condicoes = loaded_dfs.get('condicoes')
if loaded_condicoes is not None:
    # Determine description column
    desc_col_d = next(
        (c for c in ['Descrição PT_(Longa)', 'Descrição PT_(Curta)'] if c in loaded_condicoes.columns),
        None
    )
    code_col_section = 'Secção ICD-10-CM_Código'
    code_col_chapter = 'Capitulo ICD-10-CM_ Código'

    if desc_col_d and code_col_section in loaded_condicoes.columns and code_col_chapter in loaded_condicoes.columns:
        df_d = loaded_condicoes.copy()
        # Ensure codes are treated as strings and handle potential non-string types gracefully
        df_d[code_col_section] = df_d[code_col_section].apply(lambda x: str(x).strip().upper() if pd.notna(x) else '')
        df_d[code_col_chapter] = df_d[code_col_chapter].apply(lambda x: str(x).strip().upper() if pd.notna(x) else '')

        codes_section = df_d[code_col_section]
        codes_chapter = df_d[code_col_chapter]

        # S: Symptoms (R codes)
        symptom_secs = [
            'R00-R09','R10-R19','R20-R23','R25-R29',
            'R30-R39','R40-R46','R47-R49','R50-R69'
        ]
        # Check if codes_section is not empty before applying isin
        if not codes_section.empty:
            mask_s = codes_section.isin(symptom_secs)
            for term in df_d.loc[mask_s, desc_col_d].dropna():
                nt = normalize_text(term)
                if nt:
                    all_terms_by_class['S'].add(nt)
        else:
             print("Warning: codes_section is empty for S class processing.")


        # C: Causes (Chapter XX codes V00-Y99)
        # Check if codes_chapter is not empty before applying str.match
        if not codes_chapter.empty:
            mask_c = codes_chapter.str.match(r'^[VWXY]', na=False) # na=False to handle potential NaN after conversion
            for term in df_d.loc[mask_c, desc_col_d].dropna():
                nt = normalize_text(term)
                if nt:
                    all_terms_by_class['C'].add(nt)
        else:
            print("Warning: codes_chapter is empty for C class processing.")


        # D: Diagnoses/Conditions (exclude S, C, Z-, U-)
        # Ensure masks are not empty before combining
        mask_z = codes_chapter.str.startswith('Z', na=False) if not codes_chapter.empty else pd.Series([False] * len(df_d))
        mask_u = codes_chapter.str.startswith('U', na=False) if not codes_chapter.empty else pd.Series([False] * len(df_d))
        mask_s_valid = mask_s if not codes_section.empty else pd.Series([False] * len(df_d))
        mask_c_valid = mask_c if not codes_chapter.empty else pd.Series([False] * len(df_d))

        # Ensure masks have the same index as df_d
        mask_z = mask_z.reindex(df_d.index, fill_value=False)
        mask_u = mask_u.reindex(df_d.index, fill_value=False)
        mask_s_valid = mask_s_valid.reindex(df_d.index, fill_value=False)
        mask_c_valid = mask_c_valid.reindex(df_d.index, fill_value=False)


        mask_d = ~(mask_s_valid | mask_c_valid | mask_z | mask_u)
        for term in df_d.loc[mask_d, desc_col_d].dropna():
            nt = normalize_text(term)
            if nt:
                all_terms_by_class['D'].add(nt)
    else:
        print("Warning: Missing required columns in 'condicoes' for D/S/C extraction.")
else:
    print("Warning: 'condicoes' dataframe not loaded. Cannot process D, S, or C.")

# --- Process medications for M & DS ---
loaded_Medicamentos = loaded_dfs.get('medicamentos')
med_name_col = 'Nome do medicamento'
substance_col = 'Substância Ativa/DCI'
dosage_col = 'Dosagem'

if loaded_Medicamentos is not None:
    # Fill missing and ensure string type
    for col in [med_name_col, substance_col, dosage_col]:
        if col in loaded_Medicamentos.columns:
            loaded_Medicamentos[col] = (
                loaded_Medicamentos[col].fillna('')
                .astype(str)
            )
        else:
            print(f"Warning: '{col}' not found in medicamentos dataframe.")

    # Extract terms
    # Ensure columns exist before iterating
    if all(col in loaded_Medicamentos.columns for col in [med_name_col, substance_col, dosage_col]):
        for _, row in loaded_Medicamentos.iterrows():
            med_name = normalize_text(row.get(med_name_col, '')).strip()
            substance = normalize_text(row.get(substance_col, '')).strip()
            dosage = normalize_text(row.get(dosage_col, '')).strip()

            med_name_lc = med_name.lower()
            substance_lc = substance.lower()

            # M class: if substance is substring of med_name, add only substance; else add both
            if substance_lc and substance_lc in med_name_lc:
                if substance: # Ensure substance is not empty after normalization
                    all_terms_by_class['M'].add(substance)
            else:
                if med_name:
                    all_terms_by_class['M'].add(med_name)
                if substance:
                    all_terms_by_class['M'].add(substance)

            # DS class: collect raw dosage strings
            if dosage:
                all_terms_by_class['DS'].add(dosage)
    else:
        print("Warning: Not all required columns found in 'medicamentos'. Cannot process M or DS.")
else:
    print("Warning: 'medicamentos' dataframe not loaded. Cannot process M or DS.")

# --- Process procedures for PM ---
loaded_proc = loaded_dfs.get('procedimentos')
if loaded_proc is not None:
    proc_col = 'Descrição Linguagem Corrente'
    if proc_col in loaded_proc.columns:
        for term in loaded_proc[proc_col].dropna():
            nt = normalize_text(term)
            if nt:
                all_terms_by_class['PM'].add(nt)
    else:
        print(f"Warning: '{proc_col}' not found in procedimentos dataframe.")
else:
    print("Warning: 'procedimentos' dataframe not loaded. Cannot process PM.")

# --- Process allergens for AG ---
allerg = loaded_dfs.get('allergens')
if allerg is not None:
    ag_col = 'Preferred Term (PT) PT | Termo de Preferência PT'
    if ag_col in allerg.columns:
        for term in allerg[ag_col].dropna():
            nt = normalize_text(term)
            if nt:
                all_terms_by_class['AG'].add(nt)
    else:
        print(f"Warning: '{ag_col}' not found in allergens dataframe.")
else:
    print("Warning: 'allergens' dataframe not loaded. Cannot process AG.")

# --- Summary of collected terms ---
print("\n--- Terms per Category (Before Sampling) ---")
total_collected_terms = 0
for cls, terms in all_terms_by_class.items():
    label = "dosagens" if cls=='DS' else "terms"
    print(f"  {cls}: {len(terms)} {label}")
    total_collected_terms += len(terms)

print(f"Total unique collected terms: {total_collected_terms}")

```

```python
import gc
import torch
from transformers.models.auto import AutoModelForCausalLM, AutoTokenizer
from transformers import TrainingArguments
from trl import SFTTrainer
from peft import LoraConfig, get_peft_model, PeftModel
from datasets import Dataset
import matplotlib.pyplot as plt
import textwrap
from sklearn.metrics import classification_report
from tqdm import tqdm
import os
import random

# === CONFIGURAÇÕES ===
MODEL_NAME = "google/gemma-3-1b-it"
OUTPUT_DIR = "./lora_finetuned_cpu"
os.makedirs(OUTPUT_DIR, exist_ok=True)

VALID_CLASSES = {"D", "C", "PM", "M", "DS", "AG", "S"}

# === FUNÇÕES AUXILIARES ===
PROMPT_TEMPLATE = textwrap.dedent("""\
Para o termo "{term}", retorne APENAS a classe correta:
  (D, C, PM, M, DS, AG, S)
Definições:
D  = Diagnóstico/Condição (ex: Hipertensão, Diabetes)
C  = Causa               (ex: Acidente, Infecção)
PM = Procedimento Médico (ex: Ecocardiograma)
M  = Medicamento/Substância Ativa (ex: Paracetamol)
DS = Dosagem             (ex: 500 mg)
AG = Alergénio           (ex: Pólen)
S  = Sintoma             (ex: Febre)
Exemplo de saída:
D""")

def make_example(rec):
    term = rec["term"]
    label = rec["true_class"]
    return {"text": PROMPT_TEMPLATE.format(term=term) + f"\n{label}"}

def extract_class_from_output(text):
    last_line = text.strip().split("\n")[-1].strip()
    tokens = last_line.replace(":", "").replace(",", "").split()
    for token in tokens:
        if token in VALID_CLASSES:
            return token
    return "INVALID"

def get_predictions(model, tokenizer, data, max_new_tokens=5):
    model.eval()
    predictions = []
    for item in tqdm(data, desc="🔄 Gerando previsões"):
        prompt = PROMPT_TEMPLATE.format(term=item["term"])
        inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
        with torch.no_grad():
            output = model.generate(**inputs, max_new_tokens=max_new_tokens)
        decoded = tokenizer.decode(output[0], skip_special_tokens=True)
        predicted = extract_class_from_output(decoded)
        predictions.append(predicted)
    return predictions

# === CARREGAR TOKENIZER E MODELO BASE ===
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    attn_implementation="eager",
    torch_dtype=torch.bfloat16 if torch.cuda.is_available() else torch.float32
)

# === APLICAR LORA ===
lora_config = LoraConfig(
    r=16,
    target_modules=["q_proj","k_proj","v_proj","o_proj","gate_proj","up_proj","down_proj"],
    lora_alpha=16,
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# === IMPORTANTE ===
# Você deve definir `train_data` e `evaluation_data_small` antes deste script.
# Assumimos que eles já existem como listas de dicionários com chaves "term" e "true_class"

train_ds = Dataset.from_list([make_example(r) for r in train_data])
eval_ds  = Dataset.from_list([make_example(r) for r in evaluation_data_small])

# === ARGUMENTOS DE TREINAMENTO ===
training_args = TrainingArguments(
    output_dir=OUTPUT_DIR,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    warmup_steps=3,
    max_steps=40,
    learning_rate=2e-4,
    logging_steps=5,
    save_strategy="steps",
    save_steps=20,
    report_to="none"
)

# === TREINAMENTO COM SFTTrainer ===
trainer = SFTTrainer(
    model=model,
    train_dataset=train_ds,
    eval_dataset=eval_ds,
    args=training_args
)

trainer.train()

# === SALVAR MODELO E TOKENIZER ===
trainer.model.save_pretrained(OUTPUT_DIR)
tokenizer.save_pretrained(OUTPUT_DIR)

# === GRÁFICO DE PERDA ===
log_history = trainer.state.log_history
training_logs = [log for log in log_history if 'loss' in log]

if training_logs:
    steps = [log['step'] for log in training_logs]
    losses = [log['loss'] for log in training_logs]

    plt.figure(figsize=(10, 6))
    plt.plot(steps, losses, marker='o', linestyle='-')
    plt.title('Perda (Loss) de Treino ao Longo dos Passos')
    plt.xlabel('Passos de Treino')
    plt.ylabel('Perda (Loss)')
    plt.grid(True)
    plt.savefig(os.path.join(OUTPUT_DIR, 'training_loss_graph.png'))
    plt.show()

# === AVALIAÇÃO FINAL ===
print("\n🔍 Avaliando modelos BASE e FINE-TUNED...")

# Preparar dados
test_items = [{"term": e["term"]} for e in evaluation_data_small]
true_labels = [e["true_class"] for e in evaluation_data_small]

# Modelo BASE
base_model = AutoModelForCausalLM.from_pretrained(MODEL_NAME).to(model.device)
base_preds = get_predictions(base_model, tokenizer, test_items)

# Modelo FINE-TUNED
finetuned_model = AutoModelForCausalLM.from_pretrained(OUTPUT_DIR).to(model.device)
finetuned_model = PeftModel.from_pretrained(finetuned_model, OUTPUT_DIR).to(model.device)
finetuned_preds = get_predictions(finetuned_model, tokenizer, test_items)

# Filtrar INVALID
def filter_invalid(preds, labels):
    filtered_preds, filtered_labels = [], []
    for p, l in zip(preds, labels):
        if p in VALID_CLASSES:
            filtered_preds.append(p)
            filtered_labels.append(l)
    return filtered_preds, filtered_labels

print("\n📊 Métricas - Modelo BASE")
base_preds_clean, labels_clean = filter_invalid(base_preds, true_labels)
print(classification_report(labels_clean, base_preds_clean, zero_division=0))

print("\n📊 Métricas - Modelo FINE-TUNED")
finetuned_preds_clean, finetuned_labels_clean = filter_invalid(finetuned_preds, true_labels)
print(classification_report(finetuned_labels_clean, finetuned_preds_clean, zero_division=0))

# === LIMPEZA FINAL ===
del model, trainer, tokenizer
gc.collect()
if torch.cuda.is_available():
    torch.cuda.empty_cache()
```

```python
import gc
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments # TrainingArguments not strictly needed here, but kept for context
from trl import SFTTrainer # SFTTrainer not strictly needed here, but kept for context
from peft import LoraConfig, get_peft_model, PeftModel # Added PeftModel import
from datasets import Dataset # Dataset not strictly needed here, but kept for context
import pandas as pd
import random
from collections import Counter
import math
import textwrap
import os
from sklearn.model_selection import train_test_split

# --- (1) General Configurations (Ensure these are defined or imported if not) ---
MODEL_NAME        = "google/gemma-3-4b-it"
OUTPUT_DIR        = "./lora_finetuned_cpu" # This is where your fine-tuned model is saved

# Dummy values for evaluation_data_small, full_data_for_split, TEMPERATURE, TOP_P, TOP_K, MAX_TOKENS, MIN_P, TRAIN_TEST_RATIO
# You should replace these with your actual data/config if they are loaded from other files.
try:
    if 'evaluation_data_small' not in locals():
        raise NameError
    if 'full_data_for_split' not in locals():
        raise NameError
except NameError:
    print("Warning: Dummy data/config values being used. Ensure your actual data/config are loaded.")
    sample_terms = ["febre", "paracetamol", "hipertensão", "acidente", "ecocardiograma", "pólen", "500 mg"]
    sample_labels = ["S", "M", "D", "C", "PM", "AG", "DS"]
    evaluation_data_small = [{'term': random.choice(sample_terms), 'true_class': random.choice(sample_labels)} for _ in range(100)]
    full_data_for_split = evaluation_data_small * 5 # Just an example
    TEMPERATURE = 0.7
    TOP_P = 0.9
    TOP_K = 50
    # MAX_TOKENS will be set specifically for 2 tokens below. This initial MAX_TOKENS is overwritten.
    MIN_P = 0.0
    TRAIN_TEST_RATIO = 0.8


# --- (2) Data Splitting (from your original code) ---
USE_SMALL_DATASET   = True # True = small, False = full
SMALL_TRAIN_RATIO   = 0.8  # split de treino no small dataset

if USE_SMALL_DATASET:
    df = pd.DataFrame(evaluation_data_small)
    train_df, test_df = train_test_split(
        df, train_size=SMALL_TRAIN_RATIO,
        random_state=42, stratify=df["true_class"]
    )
else:
    df = pd.DataFrame(full_data_for_split)
    strat = df["true_class"] if df["true_class"].nunique()>1 else None
    train_df, test_df = train_test_split(
        df, train_size=TRAIN_TEST_RATIO,
        random_state=42, stratify=strat
    )

train_data = train_df.to_dict("records")
test_data  = test_df.to_dict("records")

# --- (3) Prompt Template (from your original code) ---
PROMPT_TEMPLATE = textwrap.dedent("""
Para o termo "{term}", retorne APENAS a classe correta:
  (D, C, PM, M, DS, AG, S)
Definições:
D  = Diagnóstico/Condição (ex: Hipertensão, Diabetes)
C  = Causa               (ex: Acidente, Infecção)
PM = Procedimento Médico (ex: Ecocardiograma)
M  = Medicamento/Substância Ativa (ex: Paracetamol)
DS = Dosagem             (ex: 500 mg)
AG = Alergénio           (ex: Pólen)
S  = Sintoma             (ex: Febre)
Exemplo de saída:
D
""").strip()

# Helper function to prepare prompt for inference
def prepare_inference_prompt(term):
    # This will be the input to the model for evaluation, WITHOUT the true label
    return PROMPT_TEMPLATE.format(term=term)


# --- (4) Modified HFCompletionAdapter to limit tokens ---
class HFCompletionAdapter:
    def __init__(self, model, tokenizer, device="cpu", max_new_tokens_limit=2):
        self.model = model.to(device)
        self.tokenizer = tokenizer
        self.device = device
        self.default_gen_kwargs = {
            "temperature": TEMPERATURE,
            "top_p":       TOP_P,
            "top_k":       TOP_K,
            "max_new_tokens": max_new_tokens_limit, # Use max_new_tokens for generation
            "min_p":       MIN_P,
            "do_sample":   False if TEMPERATURE == 0 else True, # Use sampling if temp > 0
            "num_beams":   1, # For direct generation of the class
            "pad_token_id": tokenizer.eos_token_id, # Or tokenizer.pad_token_id
            "eos_token_id": tokenizer.eos_token_id,
        }
        # Add pad token if missing
        if self.tokenizer.pad_token is None:
            self.tokenizer.add_special_tokens({'pad_token': '[PAD]'})
            self.model.resize_token_embeddings(len(self.tokenizer)) # Resize model embeddings

    def create_completion(self, prompt, **override_kwargs):
        gen_kwargs = {**self.default_gen_kwargs, **override_kwargs}
        inputs = self.tokenizer(prompt, return_tensors="pt", padding=True, truncation=True).to(self.device)

        # Generate output
        output_ids = self.model.generate(
            **inputs,
            max_new_tokens=gen_kwargs["max_new_tokens"],
            temperature=gen_kwargs["temperature"],
            top_p=gen_kwargs["top_p"],
            top_k=gen_kwargs["top_k"],
            do_sample=gen_kwargs["do_sample"],
            num_beams=gen_kwargs["num_beams"],
            pad_token_id=gen_kwargs["pad_token_id"],
            eos_token_id=gen_kwargs["eos_token_id"]
        )

        # Decode only the newly generated part (after the input prompt tokens)
        generated_token_ids = output_ids[0][ inputs["input_ids"].shape[-1] : ]
        text = self.tokenizer.decode(generated_token_ids, skip_special_tokens=True)
        return {"choices":[{"text": text}]}

# --- (5) Evaluation Function ---
# (Assuming evaluate_model_classification exists or is defined below)
def evaluate_model_classification(model_adapter, dataset_to_evaluate):
    correct_predictions = 0
    total_predictions = 0
    results = []

    print(f"Evaluating {len(dataset_to_evaluate)} examples...")
    for i, rec in enumerate(dataset_to_evaluate):
        term = rec["term"]
        true_class = rec["true_class"]
        prompt = prepare_inference_prompt(term) # Use the helper function

        # Get prediction from the model adapter
        completion = model_adapter.create_completion(prompt)
        generated_text = completion["choices"][0]["text"].strip()

        # Parse the predicted class from the generated text
        predicted_class = None
        for char in generated_text:
            if char in ['D', 'C', 'PM', 'M', 'DS', 'AG', 'S']:
                predicted_class = char
                break
        # If no valid class found, assign a placeholder
        if predicted_class is None:
            predicted_class = "[UNKNOWN]" # Or some other indicator

        is_correct = (predicted_class == true_class)
        if is_correct:
            correct_predictions += 1
        total_predictions += 1

        results.append({
            "term": term,
            "true_class": true_class,
            "predicted_class": predicted_class,
            "is_correct": is_correct,
            "generated_full_text": generated_text # Keep for debugging
        })

        if (i + 1) % 10 == 0 or (i + 1) == len(dataset_to_evaluate):
            print(f"  Processed {i+1}/{len(dataset_to_evaluate)} examples. Current Accuracy: {correct_predictions / total_predictions:.2f}")

    accuracy = (correct_predictions / total_predictions) * 100 if total_predictions > 0 else 0
    return accuracy, results

gc.collect()

# --- (6) Load and Evaluate BASE Model FIRST ---
print(f"\n🚀 Starting evaluation process...")
print(f"Loading BASE model for evaluation: {MODEL_NAME}...")
base_model_eval     = AutoModelForCausalLM.from_pretrained(MODEL_NAME, attn_implementation="eager")
base_tokenizer_eval = AutoTokenizer.from_pretrained(MODEL_NAME)

# Ensure pad token is set for the tokenizer if it's missing (good practice for generation)
if base_tokenizer_eval.pad_token is None:
    base_tokenizer_eval.add_special_tokens({'pad_token': '[PAD]'})
    # Resize model embeddings if tokens were added
    base_model_eval.resize_token_embeddings(len(base_tokenizer_eval))


# Initialize adapter for the base model, limiting to 2 new tokens
base_llama = HFCompletionAdapter(base_model_eval, base_tokenizer_eval, max_new_tokens_limit=2)

print("\n>>> Evaluating BASE MODEL...")
# Pass the model adapter and the dataset you want to evaluate on (e.g., test_data or evaluation_data_small)
base_accuracy, base_results = evaluate_model_classification(base_llama, evaluation_data_small)
gc.collect()

# --- (7) Load and Evaluate FINE-TUNED Model SECOND ---
print(f"\nLoading FINE-TUNED model for evaluation from {OUTPUT_DIR}...")
# First, load the base model architecture
fine_base = AutoModelForCausalLM.from_pretrained(MODEL_NAME, attn_implementation="eager")
# Then, load the PEFT adapters on top of it
fine_tuned_model_eval = PeftModel.from_pretrained(fine_base, OUTPUT_DIR)
# Load the tokenizer from the fine-tuned directory (it should be the same as base, but good practice)
fine_tuned_tokenizer_eval = AutoTokenizer.from_pretrained(OUTPUT_DIR)

# Ensure pad token is set for the tokenizer
if fine_tuned_tokenizer_eval.pad_token is None:
    fine_tuned_tokenizer_eval.add_special_tokens({'pad_token': '[PAD]'})
    fine_tuned_model_eval.resize_token_embeddings(len(fine_tuned_tokenizer_eval)) # Resize PEFT model embeddings

# Initialize adapter for the fine-tuned model, limiting to 2 new tokens
fine_llama = HFCompletionAdapter(fine_tuned_model_eval, fine_tuned_tokenizer_eval, max_new_tokens_limit=2)

print("\n>>> Evaluating FINE-TUNED MODEL...")
fine_tuned_accuracy, fine_tuned_results = evaluate_model_classification(fine_llama, evaluation_data_small)
gc.collect()

# --- (8) Summary ---
print("\n=== PERFORMANCE SUMMARY ===")
print(f"Base model accuracy:       {base_accuracy:.2f}%")
print(f"Fine-tuned model accuracy: {fine_tuned_accuracy:.2f}%")

print("\nEvaluation complete! ✨")
```
